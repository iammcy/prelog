# TSNE和PCA可视化

```python
from sklearn.manifold import TSNE
from sklearn.datasets import load_iris,load_digits
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import os

def TSNE_and_PCA(features, labels):
    features = features.detach().cpu().numpy()
    labels = labels.cpu().numpy()
    X_tsne = TSNE(n_components=2, random_state=33).fit_transform(features)
    X_pca = PCA(n_components=2).fit_transform(features)

    ckpt_dir="images"
    if not os.path.exists(ckpt_dir):
        os.makedirs(ckpt_dir)

    plt.figure(figsize=(10, 5))
    plt.subplot(121)
    plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, label="t-SNE", s=8)
    plt.legend()
    plt.subplot(122)
    plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, label="PCA", s=8)
    plt.legend()
    plt.savefig('images/digits_tsne-pca.png', dpi=120)
    plt.show()
    
# TSNE_and_PCA(node_features, node_labels)
```
