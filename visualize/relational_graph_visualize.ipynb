# 声明 Notebook 类型

# 关系图可视化

```python
graph_name = "dblpv7"
repulsion_size = 800000
```

## 导入 PyG 数据集（可替换）

```python
from Net2Graph import Net2Graph
graph = Net2Graph("data/" + graph_name +".mat")
print(graph)
```

## 基于 PyG 数据集的可视化

```python
from pyecharts import options as opts
from pyecharts.charts import Graph
import numpy as np

# 节点定义
"""
[{"name": "结点1", "symbolSize": 10, "category": 0}]
"""
nodes = [opts.GraphNode(name=str(idx), category=label.item()) for idx, label in zip(range(graph.num_nodes), graph.y)]

# 边定义
links = [opts.GraphLink(source=s.item(), target=t.item()) for s, t in zip(graph.edge_index[0], graph.edge_index[1])]

# 节点类别定义
categories = []
classes = np.unique(graph.y)
for i in classes:
    categories.append(opts.GraphCategory(name=str(i)))

# 图表定义
c = (
    Graph()
    .add("", nodes, links, categories, repulsion=repulsion_size)
    .set_global_opts(title_opts=opts.TitleOpts(title=graph_name))
)

c.load_javascript()
```

# Cell 内显示

```python
c.render_notebook()
```

# HTML 内显示

```python
c.render(graph_name + ".html")
```
